```python
from google.colab import files
uploaded = files.upload()  # Choose your model.pkl file when prompted
```

```python
import pickle

file = 'model.pkl'  # use the exact name of the uploaded file
load_model = pickle.load(open(file, 'rb'))
```

```python
import ipywidgets as widgets
from IPython.display import display

preg = widgets.FloatText(description='Pregnancy')
glu = widgets.FloatText(description='Glucose')
bp = widgets.FloatText(description='BP')
skin = widgets.FloatText(description='Skin')
ins = widgets.FloatText(description='Insulin')
bmi = widgets.FloatText(description='BMI')
ped = widgets.FloatText(description='DPF')
age = widgets.FloatText(description='Age')

button = widgets.Button(description="Predict")

def on_click(b):
    value = [[preg.value, glu.value, bp.value, skin.value, ins.value, bmi.value, ped.value, age.value]]
    preds = load_model.predict(value)
    print("Predicted value:", preds[0])

button.on_click(on_click)

# Display all widgets
display(preg, glu, bp, skin, ins, bmi, ped, age, button)
```
